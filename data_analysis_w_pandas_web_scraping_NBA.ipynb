# URL to the webpage "2021-22 NBA Player Stats: Per Game"

```python
# The Building Blocks
year = '2022'
url_link = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'

# Combining the URL + year strings together
url = url_link.format(year)
url
```

```python
# or simply write the url

url_ = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'
url_
```

# use pandas to read html

```python
import pandas as pd
```

```python
df = pd.read_html(url, header=0)
df
```

```python
# select the first table

df[0]
```

```python
df22=df[0]
```

```python
# Inside the table we have some non-numerical rows, exactly like the header and we don't need them.

df22[df22.Pos=='Pos']
```

```python
len(df22[df22.Pos=='Pos'])
```

```python
# we will drop these unnecessary rows

data = df22.drop(df22[df22.Pos=='Pos'].index)
data
```
